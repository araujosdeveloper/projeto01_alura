### Vamos agora analizar as tendências de imigração do Brasil em um determinado período

Vamos agora tratar só com o pais Brasil, vamos mudar o index e começar a criar nosso DataFrame específico

```python
df.set_index('País', inplace=True)
```

Obs:

set_index = essa função consgue mudar o index

inplace=True = essa função executa diretamente no DataFrame df sem a necessidade de criar um novo DataFrame

Vamos criar uma variavel para resumir o intervalo de anos ( 1980 a 2013 )

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

Vamos agora pegar somente as propriedades do Brasil

```python
brasil = df.loc['Brasil', anos]
```

Obs:

loc = essa propriedade consegue pegar rotulos específicos dentro de um DataFrame

```python
brasil
```

Vamos agora tranformar esses dados em um DataFreme

Criar um dicionário e atribuir nomes a cada uma dessas colunas

```python
brasil_dict = {'anos': brasil.index.tolist(), 'imigrantes' : brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

Vamos agora criar nosso primeiro grafico pra entendermos as tendencias de migração no Brasil

Vamos usar a biblioteca Matplotlib ( modulo pyplot )

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4)) # dimencionando o tamanho do grafico, onde 8 seria a largura e 4 a altura ( em polegadas )
plt.plot(dados_brasil['anos'], dados_brasil['imigrantes'])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Anos')
plt.ylabel('Numero de Imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.show()
```
